```python
import pandas as pd
import random

dataSet = pd.read_excel('/content/drive/MyDrive/혼자하는거/namelist.xlsx')
dataSet
```

```python
gender = input("당신의 성별은 무엇인가요? (남자: m 또는 여자: f): ")

# 성별에 따라 랜덤한 이름 생성
if gender == "m":
    name = random.choice(dataSet['일본계 성'] + dataSet['일본계 남자 이름'])
elif gender == "f":
    name = random.choice(dataSet['일본계 성'] + " " + dataSet['일본계 여자 이름'])
else:
    print("잘못된 접근입니다.")

print("당신의 일본식 이름은:", name)
```
